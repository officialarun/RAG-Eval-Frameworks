```python
import re


def clean_formula(
    formula: str
) -> str:

    # Remove equation numbering
    formula = re.sub(
        r"&&\s*\(\d+\)",
        "",
        formula
    )

    # Remove alignment markers
    formula = formula.replace(
        "&=",
        "="
    )

    formula = formula.replace(
        "&",
        ""
    )

    # Remove line breaks
    formula = formula.replace(
        "\\\\",
        " "
    )

    # Collapse tokenized words
    pattern = r'(?<=\b[A-Za-z])\s+(?=[A-Za-z]\b)'

    previous = None

    while previous != formula:

        previous = formula

        formula = re.sub(
            pattern,
            '',
            formula
        )

    # Normalize spaces
    formula = re.sub(
        r'\s+',
        ' ',
        formula
    )

    return formula.strip()
```
