```python
import re


def clean_equations(
    text: str
) -> str:

    def replace_displaystyle(
        match
    ):

        latex = match.group(1)

        replacements = {
            r"\vert": "|",
            r"\cap": "∩",
            r"\cup": "∪",
            r"\neq": "≠",
            r"\leq": "≤",
            r"\geq": "≥",
            r"\in": "∈",
            r"\notin": "∉",
        }

        for old, new in replacements.items():

            latex = latex.replace(
                old,
                new
            )

        return latex

    text = re.sub(
        r"\{\\displaystyle\s*(.*?)\}",
        replace_displaystyle,
        text,
        flags=re.DOTALL
    )

    text = re.sub(
        r"\n\s*\n\s*\n+",
        "\n\n",
        text
    )

    return text
```

```python
import re
import requests


def get_raw_wikitext(title):

    url = "https://en.wikipedia.org/w/api.php"

    params = {
        "action": "query",
        "titles": title,
        "prop": "revisions",
        "rvprop": "content",
        "format": "json"
    }

    response = requests.get(
        url,
        params=params,
        headers={
            "User-Agent": "rag-evaluation-framework"
        }
    )

    data = response.json()

    page = next(
        iter(
            data["query"]["pages"].values()
        )
    )

    return page["revisions"][0]["*"]


raw = get_raw_wikitext(
    "Bayes' theorem"
)

sections = re.findall(
    r"^==\s*(.*?)\s*==$",
    raw,
    flags=re.MULTILINE
)

print(f"Found {len(sections)} sections\n")

for section in sections:
    print(section)
```
